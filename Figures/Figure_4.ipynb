```python
#Original bar_plot function credit: Camilo Martínez M. 
#https://stackoverflow.com/questions/14270391/how-to-plot-multiple-bars-grouped
#Modified by Ritwiz Kamal

colors_toplot = ['#FFB347','#D891EF','#D1B3E6',\
                 '#9D81BA',\
                 '#77DD77','#A9A9A9','#6EA8FF','#B0E0E6','#8AC4FF','#73C6B6','#7EC8E3']

patterns = ['+','+','+','+',\
            '\\','\\','\\','\\',\
            '-','-','-','-',\
            '/','/','/','/',\
            '/','/','/','/',\
            '-','-','-','-',\
            '/','/','/','/']

pattern_indices = [(10,0),(10,1),(10,2),(10,3),\
                   (6,0),(6,1),(6,2),(6,3),\
                   (2,0),(2,1),(2,2),(2,3),\
                   (1,0),(1,1),(1,2),(1,3),\
                   (5,0),(5,1),(5,2),(5,3),\
                   (9,0),(9,1),(9,2),(9,3),\
                   (8,0),(8,1),(8,2),(8,3)]

def bar_plot(ax, data, std_devs=None, colors=None, patterns=None, pattern_indices=None,\
             total_width=0.8, single_width=1, legend=True):
    
    # Check if colors where provided, otherwhise use the default color cycle
    if colors is None:
        colors = plt.rcParams["axes.prop_cycle"].by_key()["color"]

    # Number of bars per group
    n_bars = len(data)

    # The width of a single bar
    bar_width = (total_width / n_bars)*1.4

    # List containing handles for the drawn bars, used for the legend
    bars = []

    # Build a bars_per_x dictionary depending on the number of values that are not None
    bars_per_x = {}
    for _, values_list in data.items():
        for i, value in enumerate(values_list):
            if value is not None:
                if i not in bars_per_x:
                    bars_per_x[i] = 0
                bars_per_x[i] += 1

    # Instead of using i in calculating the offset, we now use the i_per_x[x]
    i_per_x = {}

    # Iterate over all data
    for i, (name, values) in enumerate(data.items()):
        # Draw a bar for every value of that type
        for x, y in enumerate(values):
            if x not in i_per_x:
                i_per_x[x] = 0

            if y is not None:
                    
                # The offset in x direction of that bar
                x_offset = (i_per_x[x] - bars_per_x[x] / 2) * bar_width + bar_width / 2
                
                # Get standard deviation for the current bar if provided
                yerr = None
                if std_devs is not None and name in std_devs and std_devs[name][x] is not None:
                    yerr = std_devs[name][x]
                    
                # Define the error_kw dictionary to customize error bar appearance
                error_kw = {
                    'ecolor': 'red',
                    'capsize': 3.5,                       # Size of the error bar caps
                    'elinewidth': 1.5,                  # Error bar line width
                    'capthick': 1.5                     # Error bar cap thickness
                }
                
                hatch = None
                if patterns and pattern_indices and (i, x) in pattern_indices:
                    hatch = patterns[pattern_indices.index((i, x)) % len(patterns)]
                    
                # Draw the bar with error bars
                bar = ax.bar(
                    x + x_offset, 
                    y, 
                    width=bar_width * single_width, 
                    color=colors[i % len(colors)],
                    yerr=yerr,  # Add error bars
                    error_kw=error_kw,    # Set error bar appearance
                    hatch = hatch
                )
                

                i_per_x[x] += 1

        # Add a handle to the last drawn bar, which we'll need for the legend
        bars.append(bar[0])

    # Draw legend if we need
    if legend:
        fig.legend(bars, data.keys(), loc='lower center', ncol=6, bbox_to_anchor=(0.5, -0.13), fontsize=35)
```
